```python
import numpy as np #call numpy functions via the alias 'np'
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

#Numpy
numpy has a special way of accessing data from a table/matrix.
Instead of a normal array of arrays, where you access data x[0][0], you access data using only one bracket, x[rows, cols]. Further more you can use ranges, x[:, 1] means every row, but just colomn 1. (ie. selecting a single column) While x[1, :] will give you row one but with all colmns. (ie. selecting row 1)

```python
#load data from sklearn 
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)
# uncomment to run and see output
# print(diabetes_X[1,:]) #column
# print(diabetes_X[:,1]) #row
```

```python
# how would you select the third 'feature' (column) from diabetes_X
# rember index starts at ZERO!
# add the column index after the second comma
diabetes_one = diabetes_X[:, np.newaxis, 2]
# this should output 442, 1
print(diabetes_one.shape)
```

#Numpy
What if we wanted to select only the rows between 20 and 30?
This is where we can use the range operator (x[rows, cols]), x[20:30, :], columns between 2 and 5? x[:, 2:5]. Finally we can use a negative number to select starting from the end of the array. x[:-20, :] will select all rows from 0 to #of rows - 20, and all columns. 

```python
# Split diabetes_one into a test set of size 30 and a train set of the rest of the instances
# There are 442 rows
X_train = diabetes_one[:-30]
X_test = diabetes_one[-30:]
# do the same but for the y
y_train = diabetes_y[:-30]
y_test = diabetes_y[-30:]

print('Training input variable shape:', X_train.shape)
print('Test input variable shape:', X_test.shape)
# output should be 412 and 30
```

#SKlearn
Vist https://scikit-learn.org/stable/modules/linear_model.html
And find how to make a linear Regression object.

```python
#Create the regression object
regression = linear_model.LinearRegression()
#train the model, look up the fit attribute to see what params it accepts
regression.fit(X_train, y_train)
#now score the model, using test data
score = regression.score(X_test, y_test)
print(score)
#should get ~41%
prediction = regression.predict(X_test)

```

#Matplot
Run the code block and describe what shape the predictions are?

```python
# Plot outputs
plt.scatter(X_test, y_test,  color='black', label='ground truth')
plt.scatter(X_test, prediction, color='blue', label='predicted')

plt.axis('tight')
plt.tight_layout()
plt.legend()
plt.show()
```
